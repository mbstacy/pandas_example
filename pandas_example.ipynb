```python
df.MEMBERS.unique()
import itertools
#set(itertools.chain.from_iterable(df.MEMBERS.unique()))
temp=[]
for itm in set(itertools.chain.from_iterable(df.MEMBERS.unique())):
    tempcount=df['MEMBERS'].apply(lambda x: x.count(itm)).sum()
    temp.append({'name':itm,'count':tempcount})
```

```python
member_count=pandas.DataFrame(temp)
member_count.sort_values(['count', 'name'], ascending=[0, 1])
```

```python
member_count.sort_values(['count', 'name'], ascending=[0, 1]).to_excel('my.xlsx')
```
